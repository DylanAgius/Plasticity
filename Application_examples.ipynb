# Chaboche Model

## Strain control

Load all information required to run simulation.  This first example uses Multi-component Armstrong-Frederick

```python
import Chaboche as plas
                    

"type of kinematic hardening model to use"
"if traditiona chaboche use: MAF"

kinv=['MAF',1]             

"order required C1,C2,C3,Cn..,gamma1,gamma2,gamma3,gamman.. for the MAF model"

backstress=[4400,120000,5100,117000,800,4,1000,15,1800,5000]


"if you are using isotropic hardening need to answer yes"
"order required Q,b"
isoparam=['yes',-50, 1]

"material yield stress"
sigy0=200.0

"material elastic modulus"
xmod=69000

"if you want to read turning points from file answer 'yes' if not specify max"
"and minimum values interested in and the number of cycles"

"this reads turning points from a text file (example file given)"
#readfile=['yes','loading_condition.txt']

"this reads turning points and number of cycles"
readfile=['no',0.02,-0.02,5]

"number of data points per branch"
steps=100
    
strainc=plas.straincontrol(kinv,backstress,isoparam,sigy0,xmod,readfile,steps)
strainc.Plast_Model()


            
```

Plot the total strain and stress results for all cycles

```python
strainc.plotter_totalvals()
```

Plot the mean stress relaxation results

```python
strainc.plotter_meanstress()
```

Extract data of these plots to an excel sheet

```python
strainc.dataextract()
```

# Multiplicative Armstrong-Frederick Model

## Strain control

```python
import Chaboche as plas


                    
"type of kinematic hardening model to use"
"if multiplicative model use: MAFM"
"if using MAFM, please specify the number of multiplicative backstresses being used."

"This example uses 1:"
kinv=['MAFM',1]             

"if using the MAFM model, specify the C and gamma values for the multiplicative backstress "
"at the end:C1,C2,C3,C4,Cstar..,gamma1,gamma2,gamma3,gamma4,gammastar..  "

backstress=[4400,120000,5100,117000,800,4,1000,15,1800,5000]


"if you are using isotropic hardening need to answer yes"
"order required Q,b"
iso=['yes',-50, 1]

"material yield stress"
sigy0=200.0

"material elastic modulus"
xmod=69000

"if you want to read turning points from file answer 'yes' if not specify max"
"and minimum values interested in and the number of cycles"

readfile=['no',0.02,0.002,10]

"number of steps for forward Euler integration"
steps=50
    

strainc=plas.straincontrol(kinv,backstress,iso,sigy0,xmod,readfile,steps)
strainc.Plast_Model()
```
